## Data - (preparing and loading)

Data can be almost anything in machine learning:
* Excel Spreadsheet
* Images of any kind
* Videos
* Audio
* DNA
* Text

Machine learning is a game of two parts:
1. Get data int a numerical representation.
2. Build a model to learn patterns in that numerical representation.

Let's create some known data using the linear regression formula. Use linear regression formula to make a straight line with known parameters. 

Basically: y = mx + b, where we know m and b.

```python
# Cretae *known* parameters
weight = 0.7 # m
bias = 0.3 # y-intercept

# Create 
start = 0 # x_0
end = 1 # x_f
step = 0.02 # dx

X = torch.arange(start, end, step).unsqueeze(dim=1) # x-axis in tensor (matrix) form
y = weight * X + bias # y = mx + b

print(X[:10], "\n", y[:10], "\n", len(X), "\n", len(y))
```

```python
# Create a train/test split
train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]

print(len(X_train), len(y_train), len(X_test), len(y_train))
```

### Visualize!

```python
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=None):
    plt.figure(figsize=(10,7))

    # Plot training data in blue
    plt.scatter(train_data, train_labels, c="b", s=4, label="Training Data")

    # Plot testing data in green
    plt.scatter(test_data, test_labels, c="g", s=4, label="Testing Data")

    if predictions is not None:
        # Plot the predictions if they exist
        plt.scatter(test_data, predictions, c='r', s=4, label="Predictions")

    plt.legend(prop={"size": 14})

plot_predictions()
```

## Build Model

Our First PyTorch model!

Learn more about OOP in python.

Let's talk about what the below does. We're going to start with the inside portion and work our way outside:
* "self.weights": is a parameter that starts with random numbers and updates the random numbers to get closer to the **m** above.
* "self.bias": is a parameter that also starts with random numbers and updates the random numbers to get closer to the **b** above.
* "super().__init__()": just something that we need I guess...
* "__init__" function: obviously for object oriented programming.
* "forward" function: runs the function outlined above **y=mx+b** to solve for the weights and bias.
* "LinearRegressionModule" Class: a collection of two functions to solve for the m and b of a function based on the x and y's.

################################################

The code we've written is basically to trigger the following two things to occur:
1. Gradient Descent
2. Backpropagation

################################################

Any subclass of nn.Module needs to override forward() **this defines the forward computation of the model**. Basically if you're using nn.Module (which you probably always will be) then you need to define a forward function.

```python
# Create linear regression model class

class LinearRegressionModel(nn.Module): # almost everything in PyTorch inherits from nn.Module
    def __init__(self):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(1, # <- start with a random weight value
                                                requires_grad=True,
                                                dtype=torch.float))
        self.bias = nn.Parameter(torch.randn(1, # <- start with a random weight value
                                            requires_grad=True,
                                            dtype=torch.float))

    def forward(self, x: torch.Tensor) -> torch.Tensor: # <- "x" is the input data
        return self.weights * x + self.bias # this is the linear regression formula
```

### PyTorch Model Building Essentials

* torch.nn - contains all of the buildings for computational graphs (a nn is a computational graph)
* torch.nn.Parameter - what paramters should our model try and learn (constants of integration from initial conditions). Often a PyTorch layer from torch.nn will set these for us.
* torch.nn.Module - the base class for all nn modules, of you subclass it, you should overwrite forward.
* torch.optim - this is where the optimizers in PyTorch live, they will help with gradient descent.
* def forward() - all nn.Module subclasses require you to overwrite forward(), this method defines what happens in the forward computation. 

```python
# Find out what's "in" our model
# Check random seed
torch.manual_seed(42)

# Create an instance of the model (this is a sublass of nn.Module)
model_0 = LinearRegressionModel()

# Check out parameters
print(list(model_0.parameters()))

# List named parameters
print(model_0.state_dict())
```

```python
# Make predictions with model:
with torch.inference_mode(): # inference mode is good for predictions, does not do gradient.
    y_preds = model_0(X_test)

print(y_preds, "\n", y_test, "\n")

plot_predictions(predictions=y_preds)
# obviously predictions suck because we're using just random numbers for weights and bias.
```

## Training a Model

The whole idea of training is for a model to move from some unknown parameters (most likely random) to some **known** parameters.

In other words, the random weight matrix need to be refined into the ideal weight matrix for predictions.

This is done via the **loss** function (also called cost function or criterion).

### Things we need to train: 

* **Loss Function:** A function to measure how wrong your model's predictions are to ideal outputs. Lower is better!

* **Optimizer:** Takes into account the loss of a model and adjusts the model's parameters (e.g. weights, bias) to improve the loss function.

*  **Training Loop (Specifically for PyTorch)**

*  **Testing Loop (Specifically for PyTorch)**

```python
# Setup a loss function - L1Loss (Mean absolute error)
loss_fn = nn.L1Loss()

# Setup an optimizer - SGD (stochastic gradient descent)
optimizer_fn = torch.optim.SGD(params=model_0.parameters(), 
                               lr=0.01) # lr = learning rate = the most important hyperparameter (value data scientist sets) you can set
                                        # Especially important paramenter that needs to be understood as a data scientist.
                                        # lower learning rate = smaller optimization changes, 
                                        # larger " " = larger " "
```

### Building a training loop (and testing loop) in PyTorch

What we need for a training loop:

0. Loop through the data.
1. Forward pass (this involves data moving through our model's 'forward()' function(s)). **forward propagation**
2. Calculate the loss (compare forward pass predictions to ground truth labels).
3. Optimize zero grad.
4. Loss backward - move backwards through the network to calculate the gradients of each of the parameters of our modelwith respect to the loss - **back propagation**.
5. Optimizer step - use the optimizer to adjust our model's parameters to try and improve the loss - **gradient descent**

Something to look into later: learning rate scheduling.

```python
# An epoch is 1 loop through the data...
# Epochs are a hyperparameter.
epochs = 500

# For storing useful values!
# We did this after training, it's not inherently necessary, but it's for the science.
epoch_count = []
loss_values = []
test_loss_values = []

# Step 0: loop through the data
for epoch in range(epochs):

    ###########
    # TRAINING
    ###########
    
    # Set the model to training mode
    model_0.train() # train mode in PyTorch sets all parameters that require gradients to require gradients

    # 1. Forward pass:
    y_pred = model_0(X_train)

    # 2. Calculate the loss
    loss = loss_fn(y_pred, y_train)
    print("Loss: ", loss)

    # 3. Optimizer zero grad
    optimizer_fn.zero_grad()

    # 4. Perform back-propogation on the loss with respect to the parameters of the model - requires_grad=True
    loss.backward() 

    # 5. Step the optimizer (perform grad descent)
    optimizer_fn.step() # by default, how the optimizer will accumulate through the loop, so we have to zero themavove in step 3

    
    ###########
    # TESTING (predictions)
    ###########
    
    model_0.eval # turns off different settings in the model not needed for testing.
    with torch.inference_mode(): # turns off gradient tracking & a couple more things behind the scenes.
        
        # 1. Do the forward pass
        test_pred = model_0(X_test)

        # 2. Calculate Loss
        test_loss = loss_fn(test_pred, y_test)

    
    if epoch % 100 == 0:
        # print out what's happening
        print(f"Epoch: {epoch} | Loss: {loss} | Test loss: {test_loss}")
        print(model_0.state_dict()) # Print out model state_dict(). Remember: we should be aiming for .7 and .3
        # append our newly created lists:
        epoch_count.append(epoch)
        loss_values.append(loss)
        test_loss_values.append(test_loss)
            
```

```python
with torch.inference_mode():
    y_preds_new = model_0(X_test)

plot_predictions(predictions=y_preds_new)
```

```python
# Plot the loss curve
plt.plot(epoch_count,
         torch.tensor(loss_values).numpy(),
         label="Train Loss")
plt.plot(epoch_count,
         test_loss_values,
         label="Test Loss")
plt.title("Training and Test Loss Curves")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend()
```
